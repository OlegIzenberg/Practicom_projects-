```python
def rev_calc(row):
    
    plan = row['plan_name']
    mess = row['message_total']
    mb = row['monthly_traffic_rounded']
    mins = row['calls_min_total']
    surf_extra_mins = mins - 500
    surf_extra_mb = mb - 15360 
    surf_extra_mess = mess - 50 
    ult_extra_mins = mins - 3000
    ult_extra_mb = mb - 30720 
    ult_extra_mess = mess - 1000
    
    
    if plan == 'surf':
        plan_rev = 20 
    elif plan == 'ultimate':
        plan_rev = 70            
    else:
        plan_rev = 0
        

    
    if plan == 'surf' and surf_extra_mins > 0:
        extra_min_rev = surf_extra_mins * 0.03
    elif plan =='ultimate' and ult_extra_mins > 0:
        extra_min_rev = ult_extra_mins *0.01           
    else: 
        extra_min_rev = 0 
        
    
    if plan == 'surf' and surf_extra_mb > 0:
        extra_mb_rev = surf_extra_mb * (10/1024)
    elif plan == 'ultimate' and ult_extra_mb > 0:
        extra_mb_rev = ult_extra_mb * (10/1024)           
    else:
        extra_mb_rev = 0 
        
            
    if plan == 'surf' and surf_extra_mess > 0: 
        extra_mess_rev = surf_extra_mess * 0.03
    elif plan == 'ultimate' and ult_extra_mess > 0:
        extra_mess_rev = ult_extra_mess * 0.01                    
    else: 
        extra_mess_rev = 0 
        
    

    total_cost = plan_rev + extra_min_rev + extra_mb_rev + extra_mess_rev 
    return total_cost            
```

```python
merged_df4['total_rev'] = merged_df4.apply(rev_calc,axis=1)
```

## Test statistical hypotheses

<u>Hypothesis that is tested:</u>

* H0: Average of revenue from users of the Ultimate and Surf calling plans differs

* H1: We don't have enough evidence to say that average revenue from users of the Ultimate and Surf calling plans differs

```python
# Test the hypotheses
merged_df4_rev = merged_df4.query('plan_name != 0')
merged_df4_rev
```

```python
merged_df4_rev.query('total_rev > 70')
```

```python
fig, ax = plt.subplots(figsize=(8,7))
ax = sns.histplot(merged_df4_rev, x="total_rev", hue="plan_name")
```

```python
merged_df4_rev.groupby(['plan_name'])['total_rev'].median()
```

```python
merged_df4_rev_surf = merged_df4_rev.query('plan_name == "surf"')
merged_df4_rev_ultimate = merged_df4_rev.query('plan_name == "ultimate"')
```

```python
sample1=merged_df4_rev_surf['total_rev']
sample2=merged_df4_rev_ultimate['total_rev']
```

```python
sample1.isnull().sum()
sample2.isnull().sum()
```

```python
sample1.describe()
sample1.var()

sample2.describe()
sample2.var()
```

```python
st.ttest_ind(sample1, sample2)
```

```python
p_value = stats.ttest_ind(sample1, sample2).pvalue
p_value
```

```python
p_value=p_value
alpha=0.05
if p_value<alpha:
    print('Fail to Reject H0: average of sample 1 is not equal to average of sample 2')
else:
    print( "Reject H0: We don't have enough evidence to say that averages of sample 1 \
  and sample 2 are not diffrent")
```
