These are cars with a small model age gap. Not too much, but still. We can choose 20 as a threshold instead of 15 and get 800 rows. What is important here, is that we do not consider models that have 50 years age gap (e.g. chevrolet camaro).


The next question is how to make the replacement for these models? It's quite simple. I'll demonstrate it with the **olddf** dataframe: </div>


````python
# Reviewer's code

print('Nan-s Before: ', olddf.model_year.isna().sum())

for model in ll:
    # find a median for each of these models
    median_value = olddf[olddf['model'] == model]['model_year'].median()
    
    # find the rows with these models and use a simple fillna
    olddf.loc[olddf['model']==model, 'model_year'] = olddf.loc[olddf['model']==model, 'model_year'].fillna(median_value)
    
    
print('Nan-s After: ', olddf.model_year.isna().sum())    
````
